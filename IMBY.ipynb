```python
import zipfile

# Function to extract a zip file
def extract_zip(zip_path, extract_to):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)

# Iterate over each zip file in the emotion_data folder
for file_name in files:
    if file_name.endswith('.zip'):
        zip_path = os.path.join(Source, file_name)
        extract_to = os.path.join(Destination, file_name.replace('.zip', ''))

        # Create a directory to extract the files
        if not os.path.exists(extract_to):
            os.makedirs(extract_to)

        # Extract the zip file
        extract_zip(zip_path, extract_to)
        print(f'Extracted {file_name} to {extract_to}')

        # Process the extracted files
        extracted_files = os.listdir(extract_to)
        for extracted_file in extracted_files:
            file_path = os.path.join(extract_to, extracted_file)
            print(f'Processing file: {file_path}')
```

```python
from google.colab import files
import pandas as pd

from sklearn.preprocessing import StandardScaler

def preprocess_file(file_path):
    # Load the file
    df = pd.read_csv(file_path)

    # Drop rows with missing values
    df = df.dropna()

    # Convert column names to lowercase
    df.columns = [col.lower() for col in df.columns]




    # Save the processed file
    processed_file_path = file_path.replace('.csv', '_processed.csv')
    df.to_csv(processed_file_path, index=False)

    return processed_file_path
```

```python
#Calculatethe mean emotion for each candidate across the 88 time frames
import numpy as np
os.makedirs('/content/IMBY_data/Candidate_Emotion_Analysis', exist_ok=True)
cea_candidate_no=[x for x in range(1,11)]
cea_weighted_mean=[]
cea_closest_emotion=[]
cea_dominant_emotion=[]
candidate_names = {
    1.0: 'Alice',
    2.0: 'Bob',
    3.0: 'Charlie',
    4.0: 'David',
    5.0: 'Eva',
    6.0: 'Frank',
    7.0: 'Grace',
    8.0: 'Hannah',
    9.0: 'Isaac',
    10.0: 'Jack'
}
for i in range(1,11):
    ed=pd.read_csv(f'/content/IMBY_data/emotion_data/emotion_data/{i}/emotion_processed.csv')
    ed=ed.drop(columns=['movie_id'])
    ed=ed.drop(columns=['image_seq'])
    # Calculate the mean of only the numeric columns
    numeric_cols=ed.select_dtypes(include=np.number).columns
    candidate_emotions_avg=ed[numeric_cols].mean()

    # Calculate standard deviation for each emotion
    candidate_emotions_std=ed[numeric_cols].std()


    # Count the frequency of each emotion
    emotion_frequencies = ed['dominant_emotion'].value_counts()

    #Combinefeatures
    combined_features=pd.concat([candidate_emotions_avg,candidate_emotions_std,emotion_frequencies],axis=1)
    combined_features.columns=['mean','std','count']
    print(combined_features)
    combined_features = combined_features.dropna(subset=['count'])
    weighted_mean = np.average(combined_features['mean'], weights=combined_features['count'])
    print(f'Weighted Mean: {weighted_mean}')
    cea_weighted_mean.append(weighted_mean)

    # Find the emotion with the mean closest to the weighted mean
    closest_emotion = combined_features.iloc[(combined_features['mean'] - weighted_mean).abs().argmin()].name
    print(f'Closest Emotion: {closest_emotion}')
    cea_closest_emotion.append(closest_emotion)

    dominant_emotion = combined_features.loc[combined_features['count'].idxmax()].name
    print(f'Dominant Emotion: {dominant_emotion}')
    cea_dominant_emotion.append(dominant_emotion)
    emotion_counts = ed['dominant_emotion'].value_counts()

    # Create a pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(emotion_counts, labels=emotion_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Dominant Emotions for Candidate {candidate_names[i]}')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()

Emotion_features=pd.DataFrame({'Candidate no.': cea_candidate_no,'Weighted Mean': cea_weighted_mean, 'Closest Emotion': cea_closest_emotion, 'Dominant Emotion': cea_dominant_emotion})
Emotion_features.to_csv('/content/IMBY_data/Candidate_Emotion_Analysis/Emotion_features.csv', index=False)
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
os.makedirs('/content/IMBY_data/Candidate_Gaze_Analysis', exist_ok=True)
ca_candidate_no=[x for x in range(1,11)]
ca_gaze_percentage=[]
ca_blink_percentage=[]
ca_eye_offset_mean=[]
ca_eye_offset_std=[]
for i in range(1,11):
    gd=pd.read_csv(f'/content/IMBY_data/emotion_data/emotion_data/{i}/gaze_processed.csv')

    gd=gd.drop(columns=['movie_id'])
    gd=gd.drop(columns=['image_seq'])
    # print(gd)
    gaze_percentage = (gd['gaze'].sum() / len(gd)) * 100

    # Calculate blink percentage
    blink_percentage = (gd['blink'].sum() / len(gd)) * 100

    # Calculate mean and standard deviation of eye offset
    eye_offset_mean = gd['eye_offset'].mean()
    eye_offset_std = gd['eye_offset'].std()
    ca_gaze_percentage.append(gaze_percentage)
    ca_blink_percentage.append(blink_percentage)
    ca_eye_offset_mean.append(eye_offset_mean)
    ca_eye_offset_std.append(eye_offset_std)
    # Print the results
    print(f'Gaze Percentage of candidate {i} is {gaze_percentage}%')
    print(f'Blink Percentage of candidate {i} is {blink_percentage}%')
    print(f'Mean of Eye Offset of candidate {i} is {eye_offset_mean}')
    print(f'Standard Deviation of Eye Offset of candidate {i} is {eye_offset_std}')
    print()
    scaler = StandardScaler()
    gd[['eye_offset']] = scaler.fit_transform(gd[['eye_offset']])

    # Calculate the correlation matrix
    correlation_matrix = gd.corr()

    # Print the correlation matrix
    print("Correlation Matrix:")
    print(correlation_matrix)

    # Optionally, visualize the correlation matrix using a heatmap
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Feature Correlations')
    plt.show()
    print()
Gaze_features=pd.DataFrame({'Candidate no.': ca_candidate_no,'Gaze Percentage': ca_gaze_percentage, 'Blink Percentage': ca_blink_percentage, 'Eye Offset Mean': ca_eye_offset_mean, 'Eye Offset Std': ca_eye_offset_std})
Gaze_features.to_csv('/content/IMBY_data/Candidate_Gaze_Analysis/Gaze_features.csv', index=False)
```

High standard deviation (SD) of eye offset indicates significant variability in the position of the candidate's eyes relative to a neutral position.

In simpler terms, it means that the candidate's eyes frequently moved back and forth or side to side during the video.

By leveraging the negative correlation between gaze and eye_offset, you can create new features, visualize data relationships, select relevant features for modeling, and gain deeper insights into the behavior of candidates.

gaze and eye_offset: The correlation is -0.732313, which is a strong negative correlation. This means as the gaze value increases, the eye_offset value tends to decrease.

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt

# Create directory if not exists
os.makedirs('/content/IMBY_data/Candidate_Transcript_Analysis', exist_ok=True)

# Folder path for transcript data
folder_path_transcript = "/content/IMBY_data/transcript_data/transcript_data"

if not os.path.isdir(folder_path_transcript):
    print("Invalid folder path. Please provide a valid folder path.")
    exit()

# Sorted list of transcript files
files = sorted(os.listdir(folder_path_transcript), key=lambda x: int(x.split('.')[0]))

# Lists to store the results
cta_candidate_no = [x for x in range(1, len(files) + 1)]
cta_weighted_mean = []
cta_closest_sentiment = []
cta_dominant_sentiment = []
mean_speech_speed_list = []  # To store mean speech speed

# Iterate through each candidate's transcript file
for i, file_name in enumerate(files, start=1):

    # Read the CSV file for the current candidate
    df = pd.read_csv(f"{folder_path_transcript}/{file_name}")

    # Extract relevant columns (speech sentiments)
    speech_sentiment = {
        "positive": df["positive"],
        "negative": df["negative"],
        "neutral": df["neutral"],
        "confident": df["confident"],
        "hesitant": df["hesitant"],
        "concise": df["concise"],
        "enthusiastic": df["enthusiastic"]
    }

    # Calculate the mean of each sentiment
    sentiment_means = {key: value.mean() for key, value in speech_sentiment.items()}

    # Calculate the standard deviation for each sentiment
    sentiment_stds = {key: value.std() for key, value in speech_sentiment.items()}

    # Count the frequency of the dominant sentiment for each time frame
    df["dominant_sentiment"] = df[speech_sentiment.keys()].idxmax(axis=1)
    sentiment_frequencies = df["dominant_sentiment"].value_counts()

    # Combine features into a DataFrame for analysis
    combined_features = pd.DataFrame({
        'mean': sentiment_means,
        'std': sentiment_stds,
        'count': sentiment_frequencies
    })
    combined_features = combined_features.dropna(subset=['count'])  # Ensure valid entries

    # Calculate the weighted mean of the sentiments
    weighted_mean = np.average(combined_features['mean'], weights=combined_features['count'])
    cta_weighted_mean.append(weighted_mean)

    # Find the sentiment closest to the weighted mean
    closest_sentiment = combined_features.iloc[(combined_features['mean'] - weighted_mean).abs().argmin()].name
    cta_closest_sentiment.append(closest_sentiment)

    # Find the dominant sentiment (most frequent)
    dominant_sentiment = combined_features.loc[combined_features['count'].idxmax()].name
    cta_dominant_sentiment.append(dominant_sentiment)

    # Speech speed plot and analysis
    speech_speed1 = df["speech_speed"]
    mean_speech_speed = speech_speed1.mean()  # Calculate the mean speech speed
    mean_speech_speed_list.append(mean_speech_speed)  # Add to list

    # Plot sentiment analysis
    plt.figure(figsize=(6, 7))
    speech_sentiment_mean1 = {key: value.mean() for key, value in speech_sentiment.items()}
    plt.bar(speech_sentiment_mean1.keys(), speech_sentiment_mean1.values())
    plt.xlabel("Speech Sentiment")
    plt.title(f"Mean speech sentiment values of candidate {i}")
    plt.xticks(rotation=20)
    plt.savefig(f'/content/IMBY_data/Candidate_Transcript_Analysis/speech_sentiment_{i}.png')
    plt.show()

    # Plot speech speed analysis
    plt.plot(speech_speed1)
    plt.title(f"Speech speed of candidate {i} throughout the video")
    plt.savefig(f'/content/IMBY_data/Candidate_Transcript_Analysis/speech_speed_{i}.png')
    plt.show()

# Create a DataFrame to store the results similar to the first code
Transcript_features = pd.DataFrame({
    'Candidate no.': cta_candidate_no,
    'Weighted Mean Sentiment': cta_weighted_mean,
    'Closest Sentiment': cta_closest_sentiment,
    'Dominant Sentiment': cta_dominant_sentiment,
    'Mean Speech Speed': mean_speech_speed_list  # Add mean speech speed column
})

# Save to CSV file
Transcript_features.to_csv('/content/IMBY_data/Candidate_Transcript_Analysis/Transcript_features.csv', index=False)

print("Transcript analysis CSV with speech speed generated successfully.")
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import pandas as pd

# Initialize SBERT model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Example text labels (emotional and transcript data)
emotional_data = ["angry", "disgust","fear", "happy", "sad", "surprise", "neutral"]
transcript_data = ["positive", "negative", "neutral", "confident", "hesitant","concise","enthusiastic"]

# Convert emotional data to vectors
emotional_vectors = model.encode(emotional_data)

# Convert transcript data to vectors
transcript_vectors = model.encode(transcript_data)

# Combine emotional and transcript vectors into a DataFrame
emotional_df = pd.DataFrame(emotional_vectors, columns=[f'emotion_{i}' for i in range(emotional_vectors.shape[1])])
transcript_df = pd.DataFrame(transcript_vectors, columns=[f'transcript_{i}' for i in range(transcript_vectors.shape[1])])

# Add the labels to the DataFrame for reference
emotional_df['label'] = emotional_data
transcript_df['label'] = transcript_data

print("Emotional Vectors:")
print(emotional_df)
print("\nTranscript Vectors:")
print(transcript_df)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Sample Final_Analysis DataFrame
# Ensure you have the actual DataFrame as per your context
Final_Analysis = Analysis[['Candidate no.','Dominant Emotion', 'Dominant Sentiment','Mean Speech Speed','Gaze Percentage',  'Rating']]

# Step 1: Encode 'Dominant Emotion'
emotion_encoder = OneHotEncoder(sparse=False)
emotion_vectors_array = emotion_encoder.fit_transform(Final_Analysis[['Dominant Emotion']])

# Step 2: Apply PCA to 'Dominant Emotion'
emotion_pca = PCA(n_components=1)
reduced_emotion_embeddings = emotion_pca.fit_transform(emotion_vectors_array)

# Step 3: Encode 'Dominant Sentiment'
sentiment_encoder = OneHotEncoder(sparse=False)
sentiment_vectors_array = sentiment_encoder.fit_transform(Final_Analysis[['Dominant Sentiment']])

# Step 4: Apply PCA to 'Dominant Sentiment'
sentiment_pca = PCA(n_components=1)
reduced_sentiment_embeddings = sentiment_pca.fit_transform(sentiment_vectors_array)

# Step 5: Combine reduced embeddings with numerical data
# First extract Candidate no. and then combine
candidate_no = Final_Analysis[['Candidate no.']].values
combined_data = np.hstack((candidate_no, reduced_emotion_embeddings, reduced_sentiment_embeddings,
                            Final_Analysis[['Mean Speech Speed', 'Gaze Percentage',  'Rating']].values))

# Step 6: Create a new DataFrame with combined data
combined_df = pd.DataFrame(combined_data, columns=['Candidate no.', 'Emotion_Embedding', 'Sentiment_Embedding',
                                                   'Mean Speech Speed', 'Gaze Percentage', 'Rating'])

print(combined_df)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Your data
final_data = combined_df[['Emotion_Embedding', 'Sentiment_Embedding',
                           'Mean Speech Speed', 'Gaze Percentage',
                           'Rating']].values

# Step 1: Determine the optimal number of clusters using the Elbow Method
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(final_data)
    sse.append(kmeans.inertia_)

# Step 2: Plot the Elbow curve
plt.figure(figsize=(8, 6))
plt.plot(k_range, sse, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of squared distances (SSE)')
plt.title('Elbow Method For Optimal k')
plt.show()

# Step 3: Choose the optimal number of clusters
optimal_k = 3  # Adjust based on the Elbow plot
kmeans = KMeans(n_clusters=optimal_k, random_state=42)

# Step 4: Fit the model and predict clusters
combined_df['cluster'] = kmeans.fit_predict(final_data)

numeric_columns = ['Emotion_Embedding', 'Sentiment_Embedding',
                   'Mean Speech Speed', 'Gaze Percentage',
                   'Rating']

# Calculate the mean for each cluster
cluster_analysis = combined_df.groupby('cluster')[numeric_columns].mean()
print(cluster_analysis)

# Step 6: Define recruitment criteria based on cluster analysis
combined_df['recruit'] = combined_df['cluster'].apply(lambda x: 1 if x == 0 else 0)  # Example criterion
```
